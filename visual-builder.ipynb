```python
#RETURNS: CSV file clustered data in adequate Clustergrammer2 format
#'visual-.......-.csv'

#input quantity of data to visualize 
sample_size = 2500#CHANGE THIIIIIIIIIIIIIIIIIIIIIIIS

#--------
import pandas as pd
import numpy as np
import os
```

```python
#WELCOME TO SELFIE INSIGHT!

#Please input the number of layers you selected in in the kprototypes file!
    
number_of_layers = 7

#--------------

for cuy in range(2,number_of_layers):
        #INPUT CLUSTERED DATA
        data_original = pd.read_csv('/Users/joanpeluza/Desktop/EVOLUTION/kp-c'+str(cuy)+'-s10000/kp-c'+str(cuy)+'-s10000-centroids.csv')
        data = data_original.copy(deep=True).tail(sample_size)


        #save subset
        name_file = 'subset-s'+ str(sample_size)
        name_file = name_file + '.csv'
        name_file = 'visual-'+name_file

        tags=[[  ' gaze_0_y', 'Eyes','-'],
            [  ' gaze_0_z', 'Eyes','-'],
            [  ' gaze_1_y','Eyes','-'],
             [  ' gaze_1_z', 'Eyes','-'],
             [ ' gaze_angle_y', 'Eyes','-'],
             [ ' pose_Ty', 'Pose','-'],
             [ ' pose_Rx', 'Pose','-'],
             [ ' AU01_r','Eyebrows','-'],
             [  ' AU02_r', 'Eyebrows','-'],
             [ ' AU05_r', 'Eyes','-'],
             [ ' AU06_r', 'Center','Energetic Smile'],
             [ ' AU07_r', 'Eyes','Calmed Smile'],
             [ ' AU09_r', 'Center','Energetic Smile'],
             [ ' AU10_r','Mouth','Energetic Smile'],
             [  ' AU12_r', 'Mouth','Calmed Smile'],
             [ ' AU14_r', 'Mouth','Calmed Smile'],
             [ ' AU15_r', 'Mouth','Pressed Lips'],
             [ ' AU17_r', 'Mouth','Pressed Lips'],
             [ ' AU25_r', 'Mouth','Energetic Smile'],
             [ ' AU01_c','Eyebrows','-'],
             [  ' AU02_c', 'Eyebrows','-'],
             [ ' AU05_c', 'Eyes','Wide-open eyes'],
             [ ' AU06_c', 'Center','Energetic Smile'],
              [' AU07_c', 'Eyes','Calmed Smile'],
              [' AU09_c', 'Center','Energetic Smile'],
              [' AU10_c','Mouth','Energetic Smile'],
               [' AU12_c', 'Mouth','Calmed Smile'],
              [' AU14_c', 'Mouth','Calmed Smile'],
              [' AU15_c', 'Mouth','Pressed Lips'],
              [' AU17_c', 'Mouth','Pressed Lips'],
              [' AU25_c', 'Mouth','Energetic Smile'],
              [' AU28_c','Mouth','Pressed Lips']]

        #column de names
        new_names = []
        for name in data['name']:
            new_names.append('Name: '+name)
        #colum de clusters
        cluster_labels = []
        for cluster in data['cluster']:
            cluster_labels.append('Cluster: '+str(cluster))
        data.drop(['name','cluster',' confidence'],axis=1,inplace=True)


        data.insert(0,'a',new_names)
        data.insert(1,'b',cluster_labels)
        features = []
        expression = ['a','b']
        areas=['a','b']
        for feature in data.columns:
            features.append('Feature: '+feature)
            for tag in tags:
                if tag[0]==feature:
                    areas.append('Area: '+tag[1])
                    expression.append('Expression: '+tag[2])

        #aqui cambiaremos para que revise el array con las clasificaciones de features

        final_array = np.vstack((features, areas,expression,data.values))
        final_array[0,0]=''
        final_array[0,1]=''
        final_array[1,0]=''
        final_array[1,1]=''
        final_array[2,0]=''
        final_array[2,1]=''
        final_array



        df = pd.DataFrame(final_array)


        name_file = 'chosen-2,5k-'+str(cuy)+'.csv'


        df.to_csv(name_file,header =False,index = False)
```
