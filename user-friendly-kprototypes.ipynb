```python
#WELCOME TO SELFIE INSIGHT!

#This file will let you generate the clusters by layers from your data.
#Al you need to do is change the following variable, which determines how 
#many layers you want your classification to have:
    
number_of_layers = 7

#Once you change this parameter, this file will automatically as many folders as 
#number of layers you selected. Each of this folders will contain the images closer 
#to the centroids of your clusters, so you visualize the most representative expression
#contained in such cluster, files storing the values for each of the 32 dimensions and 
#and a folder containing pictures from each cluster. 

#--------------

for cuy in range(2,number_of_layers):
        n_clusters = cuy
        sample_size = 10000
        float_features_range = range(19)
        binary_features_range = range(19,32)

        #output folder: kp-c[n_clusters]-s[sample_size]

        #----------
        import pandas as pd
        import numpy as np
        import os
        from sklearn import preprocessing

        data_original = pd.read_csv('/Users/joanpeluza/Desktop/subset-chosen').tail(sample_size)
        data = data_original.copy(deep=True)
        data.drop(['name', ' confidence'],axis = 1,inplace = True)

        #create array
        array = data.values

        #columns with float data
        for i in float_features_range:
                array[:,i]= array[:,i].astype(float)


        #columns with binary data
        binary = list(binary_features_range)

        #experimental: normalize features
        #min_max_scaler = preprocessing.MinMaxScaler()
        #array= min_max_scaler.fit_transform(array)
        #TEST WITH VALUE_COUNTS: SEEMS NORMALIZED

        #data_original


        kproto = KPrototypes(n_clusters=n_clusters, verbose=2,max_iter=50) 
        clusters = kproto.fit_predict(array, categorical=binary)

        cluster_dict = []
        for c in clusters:
                cluster_dict.append(c)
        data = data_original.copy(deep =True)
        data['cluster']=cluster_dict

        #create folders for clusters
        import shutil

        folder_name = '/Users/joanpeluza/Desktop/kp-c'+str(n_clusters)+'-s'+str(sample_size)
        os.mkdir(folder_name)

        for i in range (n_clusters):
                names = data[data['cluster']==i].sample(n =10)['name']
                paths = []
                for name in names:
                        name = '/Users/joanpeluza/Desktop/selfie insight/images/'+ name.split('.')[0]+'.jpg'
                        paths.append(name)

                temp_folder_name = '/Users/joanpeluza/Desktop/g'+str(i)
                os.mkdir(temp_folder_name)
                for f in paths:
                        shutil.copy(f, temp_folder_name)

                final_folder_name = folder_name+'/g'+str(i)
                shutil.move(temp_folder_name, final_folder_name)

        folder_name

        #save results
        csv_file = 'kp-c'+str(n_clusters)+'-s'+str(sample_size)+'.csv'
        data.to_csv(csv_file,index = False)
        temp_path = '/Users/joanpeluza/Desktop/'+csv_file
        final_path = folder_name+'/'+csv_file
        shutil.move(temp_path, final_path)

        #store centroids in folder

        centroids = kproto.cluster_centroids_
        all_centroids = []
        for i in range(n_clusters):
                first = centroids[0][i]
                second = centroids[1][i]
                centroid = np.hstack((first,second))
                all_centroids.append(centroid)


        from scipy.cluster.vq import vq

            # centroids: N-dimensional array with your centroids
            # points:    N-dimensional array with your data points

        closest, distances = vq(all_centroids, array)

        names_centroids = []
        for i in closest:
                names_centroids.append(data.iloc[i][0])


        paths = []
        for name in names_centroids:
                name = '/Users/joanpeluza/Desktop/selfie insight/images/'+ name.split('.')[0]+'.jpg'
                paths.append(name)  

        temp_folder = '/Users/joanpeluza/Desktop/kp-c'+str(n_clusters)+'-s'+str(sample_size)+'/centroids'
        os.mkdir(temp_folder)
        for f,i in zip(paths,range(n_clusters)):
                shutil.copy(f, temp_folder)
                name = os.path.basename(f)
                temp_path = temp_folder +'/' +name
                final_path = temp_folder +'/'+str(i)+'.jpg'
                os.rename(temp_path,final_path)



        #save centroids data
        info_centroids = data.iloc[closest]
        info_centroids.to_csv
        centroids_file = 'kp-c'+str(n_clusters)+'-s'+str(sample_size)+'-centroids'+'.csv'
        data.to_csv(centroids_file,index = False)
        temp_path = '/Users/joanpeluza/Desktop/'+centroids_file
        final_path = folder_name+'/'+centroids_file
        shutil.move(temp_path, final_path)

```
