```python
#Now with DBSCAN 

import os
import pandas as pd
import numpy as np

data_original = pd.read_csv('/Users/joanpeluza/subset-big-no-binary.csv')
data = data_original.copy(deep=True)
data.drop(['name', ' confidence'],axis = 1,inplace = True)

```

```python

from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

```

```python
#scalate data
data = StandardScaler().fit_transform(data)
data
```

```python
#ALGORITHM
db = DBSCAN(eps=15, min_samples=3).fit(data)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)
print(labels)
```

```python
#for small subset
#eps =12, mins=3, 5 clusters, 92/228 noise
#eps =15, mins=3, 3 clusters, 35/228 noise
#for big subset
#eps =12, mins=3, 5 clusters, 1112/16089 noise

data = data_original.copy(deep = True)
data['cluster']=labels
data
```
